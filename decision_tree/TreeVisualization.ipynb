# Решающие деревья

```python
import pydotplus 
import pandas as pd
import numpy as np

from matplotlib import pyplot as plt
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

## Данные

```python
data = pd.read_csv('HR.csv')
data[:5]
```

## Строим дерево с помощью sklearn

```python
features = ['last_evaluation','number_project','average_montly_hours','time_spend_company',\
           'Work_accident','promotion_last_5years']
target = 'left'
```

```python
model = DecisionTreeClassifier(max_depth=3)
model.fit(data[features], data[target])
```

```python
plt.figure(figsize=(17,15),dpi=300)
tree.plot_tree(model,max_depth=3,feature_names=features,class_names=['stayed', 'left'],filled=True,rounded=True,\
              proportion=True,precision=2,label='all',impurity=False)
plt.show()
```
